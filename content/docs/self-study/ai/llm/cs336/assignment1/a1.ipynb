### 7.3.1 训练超参层结论

### 7.3.2 结构设计层结论

例如：

RMSNorm 对训练稳定性非常重要
Pre-Norm 比 Post-Norm 更适合作为默认结构
RoPE 对序列建模能力有明显贡献
SwiGLU 相比更简单的 SiLU FFN 带来一定收益
### 7.3.3 项目方法层结论

一个现代 decoder-only LM 的 baseline 不应只看模块堆砌，而应通过系统消融验证各组件贡献
