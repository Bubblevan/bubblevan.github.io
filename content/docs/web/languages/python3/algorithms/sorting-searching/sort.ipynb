### 冒泡排序
![](/img/python/bubble-sort.png)
其基本思想就是从左到右重复比较相邻元素，如果顺序不对就交换。
每一轮遍历结束后，当前未排序部分的最大元素会被放到末尾。
```python
temp = alist[i]
alist[i] = alist[j]
alist[j] = temp
```

````python
def bubbleSort1(alist):
    # 外层循环：固定遍历len(alist)轮（即使后面元素已排好序，仍会循环）
    for a in range(len(alist)):
        # 内层循环：每次都遍历到倒数第二个元素（重复比较已排序的部分）
        for b in range(len(alist) - 1):
            # 相邻元素比较，交换（用Python元组解包简化了交换语法）
            if alist[b] > alist[b+1]:
                alist[b], alist[b+1] = alist[b+1], alist[b]

def bubbleSort2(alist):
    # 外层循环：控制需要遍历的轮数，从最后一个元素向前递减到1
    # passnum 表示每轮需要比较的最后一个位置（因为后面的元素已排好序）
    for passnum in range(len(alist)-1, 0, -1):
        # 内层循环：遍历当前未排序的部分，比较相邻元素
        for b in range(passnum):
            # 相邻元素比较
            if alist[b] > alist[b+1]:
                # 交换（用Python元组解包简化了交换语法）
                alist[b], alist[b+1] = alist[b+1], alist[b]

def shortBubbleSort(alist):
    exchanges = True
    passnum = len(alist) - 1

    while passnum > 0 and exchanges: # 只要还需要轮数，并且上一轮还交换过没停下来，就继续排序
        exchanges = False            # 先假设这一轮不会发生交换

        for i in range(passnum):
            if alist[i] > alist[i+1]:
                exchanges = True
                alist[i], alist[i+1] = alist[i+1], alist[i]

        passnum = passnum - 1
    
alist = [54,26,93,17,77,31,44,55,20]
bubbleSort2(alist)
print(alist)
````

#### 优化冒泡
```python
for a in range(len(alist)):
    for b in range(len(alist)-1):
```
在每一轮里，最大的数已经被放到最后了，但是基础版在下一轮居然还要比较那个位置，后面排好序的部分也都还在重复比较！！！
我一开始理解错了，加上不熟悉`range`语法所以没有懂。基础版冒泡的时间复杂度是o(n^2)，不是倒三角，后者才是。
**外层**`range(len(alist)-1, 0, -1)` 表示从 `len(alist)-1` 到 `0`，步长为 `-1`。
比如`range(4, 0, -1)`，得到`4,3,2,1`，意为第一轮比较4次，第二轮比较3次，第三轮比较2次，第四轮比较1次。
而**内层**自然而然就是`passnum-1`次。
假定我们有`[5,1,4,2,8]`
- 第一轮(passnum=4，只比较4次)：
  - [1,5,4,2,8]
  - [1,4,5,2,8]
  - [1,4,2,5,8]
  - [1,4,2,5,8]
- 第二轮(passnum=3，最大的数不管)：
  - [1,4,2,5,8]
  - [1,2,4,5,8]
  - [1,2,4,5,8]
- 第三轮和第四轮无变化
### 短冒泡
上面可以看到第三轮之前就已经排好序了，但我们还是在跑，因此需要做的是提前停下来。
其最大的作用就是多了一个标记符`exchanges`用以表示这一轮有没有发生交换。

### 总结
总而言之，优化版通过缩小每轮比较范围，避免重复比较已经排好序了的尾部元素；
短冒泡则进一步增加交换标志，如果某一轮没有发生交换，说明数组已经有序，可以提前结束。
时间复杂度上，三者的最坏和平均都是O(n^2)，但短冒泡在最好情况下可以达到O(n)。
空间复杂度是O(1)，因为只需要一个额外的交换标志。
并且它是稳定排序。

### 选择排序
![](/img/python/selection-sort.png)
交换次数更少、比较index的冒泡难兄难弟。

````python
def selectionSort(alist):
    for i in range(len(alist)):
        min_index = i
        for j in range(i+1, len(alist)):
            if alist[j] < alist[min_index]:
                min_index = j
        alist[i], alist[min_index] = alist[min_index], alist[i]

alist = [54,26,93,17,77,31,44,55,20]
selectionSort(alist)
alist
````

冒泡排序会进行大量交换，而选择排序每一轮只进行一次，数字上来说就是n^2对比n-1的差别
在时间复杂度上，无论最好、平均还是最坏，选择排序都是O(n^2)，因为都要比较(n-1)+(n-2)+...+1次
无论数组是否已经有序，都会完整扫描，不像短冒泡那样可以提前结束。
空间复杂度因为没用额外数组，所以依旧O(1)。
但是稳定性就不稳定了，因为交换是可能改变相同元素的顺序的。比如说我有一个5.一个5‘，两个数值上都是5，数组则是`[5, 5', 3]`，排序完之后可能就变成`[3, 5', 5]`了。

### 插入排序
![](/img/python/insertion-sort.png)
打扑克牌整理手牌的方式就是插入排序，分为已排序和未排序两个部分分别维护。
比如说给定`[8, 3, 5, 2]`，8不管，那么第一轮currentValue就是3，position就是1。也就是假定这个currentValue是最大值插入到最大的右边position那个位置。
但很显然，3并不比8大，于是position指针向左移动，直到队列里左边的元素比currentValue小，那么这个position就是我要插入的位置。
在移动的过程中，一方面position向左走，另一方面这个已排序队列也在向右挤。`alist[position] = alist[position-1]`即元素右移，第一轮原本`[8, 3]`经过`alist[1] = alist[0]`变成`[8, 8]`。
然后position-=1变为0，不满足while循环条件跳出，然后`alist[position]=currentValue`插入得到`[3, 8|5, 2]`。


````python
def insertionSort(alist):
    for index in range(1, len(alist)): # 因为左边算作已维护，那么第0个元素就不用管他了

        currentValue = alist[index]    # 当前要插入的元素
        position = index               # 插入位置

        while position > 0 and alist[position-1] > currentValue:
            alist[position] = alist[position-1]
            position = position - 1

        alist[position] = currentValue
````

在时间复杂度方面，最坏的情况是完全逆序：`[5,4,3,2,1]`，需要移动1+2+3+...n次，即 $O(n^2)$；最好情况是已经有序，只需要遍历1次比较n次，即 $O(n)$；平均下来是 $O(n^2)$。
空间复杂度依旧 $O(1)$。
稳定性方面是稳定排序，只移动大于`currentValue`的元素，相同元素不会改变顺序。
很多高级排序，比如`Python TimSort`，是一种混合排序算法，结合了插入排序和归并排序的优点。如果数据量<20，那么插入排序就是最快的。

### 希尔排序（Shell Sort）
插入排序说是移动太慢，而希尔排序的核心思想就是先让远距离元素接近正确位置，然后再用插入排序——先粗排，再细排。插入排序也许可以看做gap=1的希尔排序，后者属于远距离预排序。

````python
def shellSort(alist):
    gap = len(alist) // 2
    while gap > 0:
        for start in range(gap):
            gapInsertionSort(alist, start, gap)

        gap = gap // 2

# 辅助函数，或者说把`insertionSort`的1换成gap即可
def gapInsertionSort(alist, start, gap):
    for i in range(start+gap, len(alist), gap):
        currentValue = alist[i]
        position = i
        while position >= gap and alist[position-gap] > currentValue:
            alist[position] = alist[position-gap]
            position = position - gap
        alist[position] = currentValue
````

时间复杂度方面，最坏依旧是O(n^2)，但是平均时间复杂度是**O(n^1.3)**。
稳定性是不稳定的，因为跨距离交换可能改变顺序。

### 归并排序
![1 个元素本身就是有序的！](/img/python/merge-sort-divide.png)
![](/img/python/merge-sort-merge.png)
第一个真正的O(nlogn)排序算法，也是分治思想(divide and conquer)的典型实践。
一句话来说，就是不断先把数组拆成小数组，再把有序的小数组合并成大数组

````python
def mergeSort(alist):
    if len(alist) > 1:
        mid = len(alist) // 2
        lefthalf = alist[:mid]
        righthalf = alist[mid:]

        mergeSort(lefthalf)
        mergeSort(righthalf)

        i = j = k = 0 
        # 3个指针分别对应左数组、右数组、原数组
        while i < len(lefthalf) and j < len(righthalf):
            if lefthalf[i] < righthalf[j]:
                alist[k] = lefthalf[i]
                i = i + 1
            elif lefthalf[i] >= righthalf[j]:
                alist[k] = righthalf[j]
                j = j + 1
            k = k + 1

        # 如果长度不一样的话，把剩下的加进去
        while i < len(lefthalf):
            alist[k] = lefthalf[i]
            i = i + 1
            k = k + 1

        while j < len(righthalf):
            alist[k] = righthalf[j]
            j = j + 1
            k = k + 1
````

时间复杂度方面，首先看divide，比如8个元素的数组，要拆3层得到1个元素的最小单位，所以是O(logn)；然后看merge，每层把所有元素合并一遍的时间复杂度是O(n)，所以总时间复杂度是O(nlogn)。
不论原数组是有序、乱序、逆序，拆分的层数永远是logn，而merge总工作量永远是n，因此它不会退化。

空间复杂度方面，merge的时候需要额外开一个临时数组来存合并后的结果（比如合并`[3, 8]`和`[4, 5]`，需要临时数组存`[3, 4, 5, 8]`），这个临时数组的最大长度就是原数组的长度，所以是O(n)。

其merge两个有序数组时，比如左边是`[2, 5]`，右边是`[2', 7]`。当左元素=右元素时，优先取左边的元素，结果是`[2, 2', 5, 7]`，所以是稳定的（如果你相等取右边的那自然就不稳定了）。

### 快速排序
![](/img/python/quick-sort.png)
说是排序算法里最经典最常用面试最爱问的算法？
总而言之吗，核心思想就是选一个基准（pivot），把数组分成两遍，小的往左，大的往右，然后递归继续分治。
```bash
                [8,3,1,7,0,10,2]
                       |
                     pivot=8
                /             \
        [3,1,7,0,2]          [10]
            |
          pivot=3
       /           \
   [1,0,2]         [7]
      |
    pivot=1
   /        \
 [0]        [2]
```
以 `alist = [54,26,93,17,77,31,44,55,20]` 为例，演示第一次分区：
1. 基准值 `pivotvalue = 54`，`leftmark=1`（26），`rightmark=8`（20）
2. 左指针移动：26≤54 → 移到2（93）（93>54，停止）
3. 右指针移动：20<54 → 停止（rightmark=8）
4. 交换leftmark(2)和rightmark(8) → 数组变成 `[54,26,20,17,77,31,44,55,93]`
5. 继续循环：
   - 左指针从3开始：17≤54 → 移到4（77）（77>54，停止）
   - 右指针从8开始：93≥54 → 移到7（55）→ 移到6（44）（44<54，停止）
6. 交换leftmark(4)和rightmark(6) → 数组变成 `[54,26,20,17,44,31,77,55,93]`
7. 继续循环：
   - 左指针从5开始：31≤54 → 移到6（77）（77>54，停止）
   - 右指针从6开始：77≥54 → 移到5（31）→ 此时rightmark(5) < leftmark(6) → 循环结束
8. 基准值归位：交换first(0)和rightmark(5) → 数组变成 `[31,26,20,17,44,54,77,55,93]`
9. 返回分割点5 → 基准值54站到了索引5的位置，左边≤54，右边≥54

接下来递归处理左边`[31,26,20,17,44]`和右边`[77,55,93]`，逻辑完全一样，直到所有元素有序。

````python
def quickSort(alist):
    quickSortHelper(alist, 0, len(alist)-1)

def quickSortHelper(alist, first, last):
    if first < last:
        # partition函数一方面将列表分为两部分，另一方面返回基准值最终的位置（分割点）
        splitpoint = partition(alist, first, last)

        quickSortHelper(alist, first, splitpoint-1)
        quickSortHelper(alist, splitpoint+1, last)

def partition(alist, first, last):
    pivotvalue = alist[first]   # 基准，默认选第0个元素

    leftmark = first + 1        # 左指针，在pivot右边找大于pivot的元素
                                # 最终目的是把小于等于基准的元素放左边
    rightmark = last            # 右指针，在pivot左边找小于pivot的元素

    done = False

    while not done:
        # 下面两个while循环主要用于描述指针移动的条件
        # 条件1：左指针始终小于右指针，不会数组越界
        # 条件2：我能找到满足基准比较标准的元素
        while leftmark <= rightmark and alist[leftmark] <= pivotvalue:
            leftmark = leftmark + 1

        while leftmark <= rightmark and alist[rightmark] >= pivotvalue:
            rightmark = rightmark - 1

        if leftmark > rightmark:
            done = True   # 分区已完成
        else:
            alist[leftmark], alist[rightmark] = alist[rightmark], alist[leftmark]
    
    # 分区完成之后，右指针的最终位置就是pivot要放的分割点
    # 循环结束时 leftmark > rightmark，leftmark 指向的是「第一个 ≥ 基准值」的元素
    alist[first], alist[rightmark] = alist[rightmark], alist[first]

    return rightmark

alist = [54,26,93,17,77,31,44,55,20]
quickSort(alist)
print(alist)
````

时间复杂度方面，最理想的情况是每次分区都能将数组一分为二，这样递归深度就是 $O(\log n)$，每次分区的时间复杂度是 $O(n)$，所以整体时间复杂度是 $O(n \log n)$。平均也是 $O(n \log n)$。
然而标准快排选择第一个元素当基准，如果数组已经有序，那么每次分区只能将数组分成一个元素和剩余元素两部分，递归深度就是 $O(n)$，时间复杂度就退化成 $O(n^2)$。
对于此，优化方案包括选择中间元素、随机元素当基准。

空间复杂度方面，递归调用会占用栈空间，最好和平均都是 $O(\log n)$，最坏是 $O(n)$。而它是原地排序不需要额外数组（但**递归栈**的空间必须算），因此比归并排序省空间。上面的归并因为`lefthalf`和`righthalf`做了数组切片创建了新数组，所以开辟了和原数组同规模的额外数组。
> 函数调用栈的大小是固定的，和要处理的数组长度无关，比如你处理 10 个元素的数组，和处理 10000 个元素的数组，「申请表」的格式、大小完全一样，只是里面填的参数（first/last）数值不同。
| 类型                | 存储内容                | 大小和n的关系       | 空间复杂度 |
|---------------------|-------------------------|---------------------|------------|
| 快排的栈帧空间      | 函数参数、返回地址等（无数组元素） | 仅和递归层数有关（logn/n） | O(logn)/O(n) |
| 归并的额外数组空间  | 数组元素（比如lefthalf = alist[:mid]） | 和数组长度n成正比（每层都要存n个元素） | O(n) |

快排是不稳定排序。

### 堆排序
先了解什么是堆：
#### 二叉堆（Binary Heap）
堆用数组表示即可：`[50, 30, 40, 10, 20, 35]`
左孩子 = 2*i+1
右孩子 = 2*i+2
父节点 = (i-1)//2
得到这样的树：
```bash
        50
       /  \
     30    40
    / \    /
  10 20  35
```
而所谓的最大堆（Max Heap），指的就是父节点>=自己点的二叉堆，最大值一定在根节点。
#### 堆排序的过程
1. 先将数组堆化（Heapify），即将数组转换为最大堆。
2. 交换根节点和最后一个节点，将最大值放到数组末尾。
3. 对根节点进行下沉操作（Sift Down），将新的根节点放到最后元素位置，将堆大小减1重新调整堆。
4. 重复步骤2和3，直到堆中只有一个节点。

````python
def HeapSort(alist):
    n = len(alist)
    # 先将数组堆化
    for i in range(n//2 - 1, -1, -1):
        heapify(alist, n, i)

    for i in range(n-1, 0, -1):
        alist[i], alist[0] = alist[0], alist[i]
        heapify(alist, i, 0)

def heapify(alist, n, i):
    largest = i
    left = 2*i + 1
    right = 2*i + 2

    if left < n and alist[left] > alist[largest]:
        largest = left
    if right < n and alist[right] > alist[largest]:
        largest = right

    if largest != i:
      alist[i], alist[largest] = alist[largest], alist[i]
      heapify(alist, n, largest)
````

时间复杂度方面，构建堆每次花 O(n)，每次取最大值花 O(logn)，所以总时间复杂度为 O(nlogn)。
堆排序是原地排序算法，空间复杂度为O(1)。
堆排序依旧不稳定。

## 1. 利用哈希表性能公式，计算平均比较次数

设装填因子为 `λ`。

常见公式是：

* **成功查找**平均比较次数
  [
  C_s=\frac12\left(1+\frac1{1-\lambda}\right)
  ]

* **失败查找**平均比较次数
  [
  C_u=\frac12\left(1+\frac1{(1-\lambda)^2}\right)
  ]

代入各装填率：

| 装填率 λ      | 成功查找平均比较次数 (C_s) | 失败查找平均比较次数 (C_u) |
| ---------- | ---------------: | ---------------: |
| 10% = 0.10 |            1.056 |            1.117 |
| 25% = 0.25 |            1.167 |            1.389 |
| 50% = 0.50 |            1.500 |            2.500 |
| 75% = 0.75 |            2.500 |            8.500 |
| 90% = 0.90 |            5.500 |           50.500 |
| 99% = 0.99 |           50.500 |         5000.500 |

### 结论

装填率越高，冲突越严重。尤其是 75% 以后，失败查找代价暴涨；到 90%、99% 时几乎已经不可接受。

## 2. 哈希表什么时候算太小了？为什么？

我认为当**装填因子超过 0.7～0.75 左右**时，哈希表通常就算太小了。

原因：

1. **冲突明显增多**
   元素太多、桶太少，多个键更容易映射到同一位置。

2. **查找和插入都变慢**
   尤其在线性探测中，会出现“聚集”现象，连续一大片位置都被占用。

3. **失败查找代价特别高**
   从第 1 题可以看出，75% 满时失败查找平均比较次数已到 8.5；90% 时直接到 50.5。

4. **扩容成本虽有，但总体值得**
   与其让每次操作都变慢，不如适时扩容保持较低负载。

所以工程上通常会在装填因子达到某个阈值时扩容，比如 0.7、0.75 或 0.8。

## 3. 修改字符串哈希函数，使其使用位置权重

普通字符串哈希常写成：

```python
def hash_string(s, table_size):
    total = 0
    for ch in s:
        total += ord(ch)
    return total % table_size
```

这个函数的问题是：`"abc"` 和 `"cba"` 得到的和一样，顺序信息丢失了。

加入**位置权重**后：

```python
def hash_string_weighted(s, table_size):
    total = 0
    for i, ch in enumerate(s):
        total += (i + 1) * ord(ch)
    return total % table_size
```

例如：

* `"abc"`
  [
  1\cdot ord('a')+2\cdot ord('b')+3\cdot ord('c')
  ]
* `"cba"`
  [
  1\cdot ord('c')+2\cdot ord('b')+3\cdot ord('a')
  ]

两者一般就不会相同。

## 4. 设计一种替代的加权方案，并说明偏差

一种替代方案可以用**平方权重**：

```python
def hash_string_square_weight(s, table_size):
    total = 0
    for i, ch in enumerate(s):
        total += ((i + 1) ** 2) * ord(ch)
    return total % table_size
```

也可以用**多项式权重**：

```python
def hash_string_poly(s, table_size, base=31):
    total = 0
    for i, ch in enumerate(s):
        total += ord(ch) * (base ** i)
    return total % table_size
```

### 这些加权函数的偏差

1. **前面字符或后面字符可能被过度放大**
   比如平方权重、指数权重会让后面字符影响过大。

2. **某些模式字符串容易扎堆**
   比如名字、单词往往前缀很像，可能导致某些桶更常被命中。

3. **模数选择会影响分布**
   如果表长和权重规律不合适，可能会出现周期性冲突。

4. **字符集分布不均**
   英文单词中某些字母本来就高频，导致哈希值分布不一定均匀。

### 总结

带权哈希比简单求和好，但仍然不是“绝对均匀”的；权重设计和表大小一起决定实际效果。

## 5. 研究完美哈希函数，并用姓名列表示例

### 完美哈希函数是什么

对于一个**固定集合**，如果哈希函数能让集合中每个键都映射到不同地址、完全无冲突，这就是完美哈希。
如果还正好映射到连续区间最小范围，则称**最小完美哈希**。

### 示例姓名表

设姓名集合为：

```text
Alice, Bob, Carol, David, Eve, Frank, Grace, Heidi
```

我构造一个简单函数：

[
h(s)=\left(\sum_{i=0}^{n-1}(i+1)^2\cdot ord(s[i])\right)\bmod 10
]

对这 8 个名字计算得到：

| 姓名    | 哈希值 |
| ----- | --: |
| Alice |   1 |
| Bob   |   2 |
| Carol |   7 |
| David |   8 |
| Eve   |   0 |
| Frank |   4 |
| Grace |   9 |
| Heidi |   6 |

这些值互不重复，所以对这组固定名字来说，它就是一个**完美哈希函数**。

### 说明

它不是对“所有字符串”都完美，只对这个固定集合完美。
这正是完美哈希常见的应用场景：**静态关键字集合**。

## 6. 随机整数列表，用各算法排序

取随机列表示例：

```text
[37, 12, 25, 9, 56, 18]
```

最终结果都应为：

```text
[9, 12, 18, 25, 37, 56]
```

### 6.1 冒泡排序

每轮把最大值冒到后面。

* 初始：`[37, 12, 25, 9, 56, 18]`
* 第1轮：`[12, 25, 9, 37, 18, 56]`
* 第2轮：`[12, 9, 25, 18, 37, 56]`
* 第3轮：`[9, 12, 18, 25, 37, 56]`
* 第4轮：`[9, 12, 18, 25, 37, 56]`
* 第5轮：`[9, 12, 18, 25, 37, 56]`

### 6.2 选择排序

每轮选最小值放到前面。

* 初始：`[37, 12, 25, 9, 56, 18]`
* 第1轮选 9：`[9, 12, 25, 37, 56, 18]`
* 第2轮选 12：`[9, 12, 25, 37, 56, 18]`
* 第3轮选 18：`[9, 12, 18, 37, 56, 25]`
* 第4轮选 25：`[9, 12, 18, 25, 56, 37]`
* 第5轮选 37：`[9, 12, 18, 25, 37, 56]`

### 6.3 插入排序

维护左边有序。

* 初始：`[37, 12, 25, 9, 56, 18]`
* 插入 12：`[12, 37, 25, 9, 56, 18]`
* 插入 25：`[12, 25, 37, 9, 56, 18]`
* 插入 9：`[9, 12, 25, 37, 56, 18]`
* 插入 56：`[9, 12, 25, 37, 56, 18]`
* 插入 18：`[9, 12, 18, 25, 37, 56]`

### 6.4 希尔排序（gap 取 3, 1）

* 初始：`[37, 12, 25, 9, 56, 18]`

`gap = 3`，分组：

* `(37, 9)` -> 排序后 `(9, 37)`
* `(12, 56)` -> 不变
* `(25, 18)` -> 排序后 `(18, 25)`

得到：`[9, 12, 18, 37, 56, 25]`

`gap = 1`，做插入排序：

* 结果：`[9, 12, 18, 25, 37, 56]`

### 6.5 归并排序

* 拆分：`[37,12,25] [9,56,18]`
* 再拆：`[37] [12,25] [9] [56,18]`
* 再拆：`[37] [12] [25] [9] [56] [18]`
* 合并：`[12,25]`
* 合并：`[37] + [12,25] -> [12,25,37]`
* 合并：`[18,56]`
* 合并：`[9] + [18,56] -> [9,18,56]`
* 最终合并：`[12,25,37] + [9,18,56] -> [9,12,18,25,37,56]`

### 6.6 快速排序（基准取首元素）

* 初始：`[37, 12, 25, 9, 56, 18]`
* pivot=37
  分区后：`[12, 25, 9, 18] 37 [56]`
* 对左边 `[12,25,9,18]`，pivot=12
  分区：`[9] 12 [25,18]`
* 对 `[25,18]`，pivot=25
  分区：`[18] 25 []`
* 最终：`[9, 12, 18, 25, 37, 56]`

## 9. 对字符列表 `['P','Y','T','H','O','N']` 排序

按字母升序应得到：

```text
['H','N','O','P','T','Y']
```

### 9.1 冒泡排序

* 初始：`[P,Y,T,H,O,N]`
* 第1轮：`[P,T,H,O,N,Y]`
* 第2轮：`[P,H,O,N,T,Y]`
* 第3轮：`[H,O,N,P,T,Y]`
* 第4轮：`[H,N,O,P,T,Y]`
* 最终：`[H,N,O,P,T,Y]`

### 9.2 选择排序

* 第1轮选最小 `H`：`[H,Y,T,P,O,N]`
* 第2轮选最小 `N`：`[H,N,T,P,O,Y]`
* 第3轮选最小 `O`：`[H,N,O,P,T,Y]`
* 后面已自然有序

### 9.3 插入排序

* 插入 Y：`[P,Y,T,H,O,N]`
* 插入 T：`[P,T,Y,H,O,N]`
* 插入 H：`[H,P,T,Y,O,N]`
* 插入 O：`[H,O,P,T,Y,N]`
* 插入 N：`[H,N,O,P,T,Y]`

### 9.4 希尔排序（gap 取 3,1）

* 初始：`[P,Y,T,H,O,N]`

`gap=3`：

* `(P,H)` -> `(H,P)`
* `(Y,O)` -> `(O,Y)`
* `(T,N)` -> `(N,T)`

得到：`[H,O,N,P,Y,T]`

`gap=1` 再插入排序：

* `"[H,N,O,P,T,Y]"`

### 9.5 归并排序

* 拆：`[P,Y,T] [H,O,N]`
* 排左：`[P,T,Y]`
* 排右：`[H,N,O]`
* 合并：`[H,N,O,P,T,Y]`

### 9.6 快速排序（基准取首元素 P）

* pivot=P
  左边：`[H,O,N]`
  右边：`[Y,T]`
* 左边再排：pivot=H -> `[] H [O,N]`
* `[O,N]` -> `[N] O []`
* 右边 `[Y,T]` -> `[T] Y []`

最终：`[H,N,O,P,T,Y]`

## 10. 设计快速排序基准值的替代策略，并分析优劣

### 替代策略：选择中间元素作为基准

例如：

```python
def quicksort_middle(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    mid = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quicksort_middle(left) + mid + quicksort_middle(right)
```

### 与“取第一个元素”为基准相比

#### 更好的情况

1. **已经有序或接近有序的数据**

   * 取第一个元素时，容易退化成最坏情况 `O(n^2)`
   * 取中间元素通常更平衡，性能更接近 `O(n log n)`

2. **逆序数据**

   * 同理，取中间元素也比取首元素更稳定

3. **分布比较均匀的数据**

   * 中间元素常常能带来更接近平衡的划分

#### 可能更差的地方

1. **额外开销略大**

   * 需要访问中间位置，某些实现还会复制左右子数组

2. **中间元素不一定真是“好基准”**

   * 它只是位置在中间，不代表数值在中间

3. **对某些特殊构造数据仍可能不理想**

   * 不能保证永远避免最坏情况

### 更合理的评价标准

比较新策略和原策略时，可以看：

1. **运行时间**
2. **比较次数**
3. **交换次数**
4. **递归深度**
5. **分区是否平衡**
6. **在已排序、逆序、随机数据上的表现**

### 总结

* 若数据常常**接近有序**，中间元素基准通常比首元素更好。
* 若只看简单实现，首元素法代码更短，但更容易退化。
* 工程上更常见的是：

  * **随机基准**
  * **三数取中**
