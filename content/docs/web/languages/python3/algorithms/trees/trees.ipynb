## 基础特性
1. **层次性**，可沿着“根→中间节点→叶节点”的路径逐层访问；
2. **节点子节点独立性**，不同父节点下的子节点彼此独立，即使名称相同，也属于不同节点，修改其一不会影响另一；
3. **叶节点唯一性**，从树根到任意叶节点的**完整路径**是唯一的，可通过该路径精准标识叶节点；
4. **子树可迁移性**，树的任意子树可整体移动到树的其他位置，仅改变子树的上层路径，不会影响子树内部的层级结构和内容；

#### 从算式到解析树
`( 3 + ( 4 * 5 ) )`拆成列表就是

分为4种规则：
- 左括号`(`：给当前节点加一个左孩子，然后往左走到这个左孩子
- 运算符`+-*/`：把当前节点的值设为这个运算符，然后加一个右孩子并往右走
- 数字：当前节点值设为该数字，回到父节点（准备读运算符了嘛）
- 右括号`)`：回到父节点（子表达式结束）


````python


def buildParseTree(fpexp):
    fplist = fpexp.split()  # 把表达式拆成token
    pStack = Stack()        # 用栈记父节点
    eTree = BinaryTree('')

    pStack.push(eTree)
    currentTree = eTree

    for i in fplist:
        if i == '(':
            currentTree.insertLeft('')
            pStack.push(currentTree)
            currentTree = currentTree.getLeftChild()

        elif i in ['+', '-', '*', '/']:
            currentTree.setRootVal(i)
            currentTree.insertRight('')
            pStack.push(currentTree)
            currentTree = currentTree.getRightChild()

        elif i == ')':
            currentTree = pStack.pop()

        elif i not in ['+', '-', '*', '/', ')']:
            try:
                currentTree.setRootVal(int(i))
                parent = pStack.pop()
                currentTree = parent

            except ValueError:
                raise ValueError("token '{}' is not a valid integer".format(i))

    return eTree

pt = buildParseTree("( ( 10 + 5 ) * 3 )")
pt.postorder()  #defined and explained in the next section
````

#### 从解析树到结果
递归思想

````python
import operator

def evaluate(parseTree):
    opers = {
        '+': operator.add,
        '-': operator.sub,
        '*': operator.mul,
        '/': operator.truediv
    }

    leftC = parseTree.getLeftChild()
    rightC = parseTree.getRightChild()

    if leftC and rightC:
        fn = opers[parseTree.getRootVal()]
        return fn(evaluate(leftC), evaluate(rightC))
    else:
        return parseTree.getRootVal()
````

### 优先队列
普通队列是FIFO，但是优先队列则是按**优先级**决定**出队顺序**。
比如说用列表去实现这样的优先队列，插入元素的时间复杂度是O(n)，再进行排序的时间复杂度是O(nlogn)。
有一说一，超级麻烦：
- 插入后还要找位置
- 或者每次取最小值都要重新排序
但实际上可以用一种称为**二叉堆**的数据结构做得更好——其能够实现出队和入队操作的时间复杂度都达到O(logn)！！
## 二叉堆
### 分类
- 小顶堆：每个节点的值都小于等于其左右子节点的值
- 大顶堆：每个节点的值都大于等于其左右子节点的值
### 实现小顶堆
#### 结构特性
完全二叉树：除了最后一层，前面每层都满。最后一层从左到右依次填。
![](/img/python/complete-binary-tree.png)
用完全二叉树可以保证节点位置的规整——能直接用列表表示。
加入一个节点在下标i：
- 左孩子在2i
- 右孩子在2i+1
- 爸爸在i//2
#### 堆序特性
除了“形状上是完全二叉树”，堆序特性要求（最小堆）满足**每个父节点都 小于等于 它的孩子**：
- 根节点一定最小
- 但整棵树不要求**完全排序**，他只要求爸爸<=儿子，对世子之争没有要求
```
        2
      /   \
     3     5
    / \
   9   6
```


````python
class BinHeap:
    def __init__(self):
        self.heapList = [0]   # heapList 用来存堆，第 0 位放个没用的占位符
        self.currentSize = 0  # currentSize 记录现在有几个元素
````

![将新元素在树中向上渗透到维持堆特性所需的位置](/img/python/percolate-up.png)
一旦新元素被添加到树中，percUp就会接管并将新元素放置在合适的位置。

````python
class BinHeap:
    def percUp(self, i):
        while i // 2 > 0:
            if self.heapList[i] < self.heapList[i // 2]:
                tmp = self.heapList[i//2]               # 获取父节点，值是tmp
                self.heapList[i//2] = self.heapList[i]  # 被插入值移到上面去
                self.heapList[i] = tmp                  # 爸爸下来
            i = i // 2

    def insert(self, k):
        self.heapList.append(k)
        self.currentSize = self.currentSize + 1
        self.percUp(self.currentSize)
````

![](/img/python/delmin.png)
因为最小堆的根节点最小，所以删最小值就是删根。但删根之后不能空着，需要做两步来**恢复堆结构**：
- 把最后一个元素搬到根
- 让它往**下沉**，恢复堆序性质

> 要理解下面的下沉方法的话，需要从下往上看这3个方法

````python
class BinHeap:
    def minChild(self, i):
        if i*2 + 1 > self.currentSize:
            return i*2        # 没有右孩子那就是左孩子
        else:
            if self.heapList[i*2] < self.heapList[i*2+1]:
                return i*2
            else:
                return i*2+1

    def percDown(self, i):
        while(i*2) <= self.currentSize:
            mc = self.minChild(i)             # mc应该是minimal character的意思？总而言之需要左右世子中更小的那个，因此自己实现
            if self.heapList[i] > self.heapList[mc]:
                self.heapList[i], self.heapList[mc] = self.heapList[mc], self.heapList[i]
            i = mc

    def delMin(self):
        retval = self.heapList[1]
        self.heapList[1] = self.heapList[self.currentSize]  # 把最后一个元素搬到根
        self.currentSize = self.currentSize - 1
        self.heapList.pop()  # 删除最后一个元素
        self.percDown(1)    # 让它往下沉，恢复堆序性质
        return retval
````

给你一个普通列表，如何直接原地整理成一个堆，而不是一个个调用insert？
![](/img/python/buildheap.png)

````python
class BinHeap:
    def buildHeap(self, alist):
        i = len(alist) // 2       # 因为最小的左叶子是2i嘛，后面的都是叶子，所以根节点只能在1~2i之间
        self.currentSize = len(alist)
        self.heapList = [0] + alist[:]
        while(i > 0):             # 从“最后一个非叶子节点”开始，一路往前下沉到叶子，直到根
            self.percDown(i)
            i = i - 1
````

## 二叉搜索树
前面用完全二叉树->最小堆/最大堆，是为了实现优先队列；
而这里是用二叉搜索树来实现字典这种**键值对结构**。
map 本质是 **用 key 找 value**。给定`{6, "yellow"}`的6和一个字典，我如何最少的查询次数来得到"yellow"？
查找时不是遍历全部，而是根据大小关系缩小范围：
- 如果目标 key 比当前小 → 去左边
- 比当前大 → 去右边
- 相等 → 找到了
这就像 **“树形版二分查找”**。根据 key 的大小关系，快速找到对应的 value。
BST 的核心性质就一句：**左边比我小，右边比我大**。
这个规则不只是根节点成立，是 **每个节点** 都成立。
![](/img/python/binary-search-tree.png)



````python
# 整颗搜索树
class BinarySearchTree:
    def __init__(self):
        self.root = None
        self.size = 0
    def length(self):
        return self.size
    def __len__(self):
        return self.size
    def __iter__(self):
        return self.root.__iter__()

# 单个节点
class TreeNode:
    def __init__(self, key, val, left=None, right=None, parent=None):
        self.key = key
        self.payload = val  # 这个名字是为了区分key，key是用来比较的，payload才是存储的值
        self.leftChild = left
        self.rightChild = right
        self.parent = parent
    def hasLeftChild(self):
        return self.leftChild
    def hasRightChild(self):
        return self.rightChild
    def isLeftChild(self):
        return self.parent and self.parent.leftChild == self
    def isRightChild(self):
        return self.parent and self.parent.righChild == self
    def isRoot(self):
        return not self.parent
    def isLeaf(self):
        return not (self.rightChild or self.leftChild)
    def hasAnyChildren(self):
        return self.rightChild
    def hasBothChildren(self):
        return self.rightChild and self.leftChild
    def replaceNodeData(self, key, value, left, right):
        self.key = key
        self.paylaod = value
        self.leftChild = left
        self.rightChild = right
        if self.hasLeftChild():
            self.leftChild.parent = self
        if self.hasRightChild():
            self.rightChild.parent = self
````

接下来是BST最基础的操作 `put`:
从根开始比较：
- 新key小于当前节点 -> 去左子树
- 新key大于当前节点 -> 去右子树
- 直到找到空位置把新节点放进去

````python
class BinarySearchTree:
    def put(self, key, val):
        if self.root:   # 树非空
            self._put(key, val, self.root)
        else:           # 如果是颗空树，直接把该节点当根
            self.root = TreeNode(key, val)
        self.size = self.size + 1   # 这里已经做了，所以_put不用再做了
    
    def _put(self, key, val, currentNode):  # 内部私有方法
        # 情况1：新key < 当前节点的key → 往左走
        if key < currentNode.key:
            if currentNode.hasLeftChild():
                self._put(key, val, currentNode.leftChild)
            else:
                currentNode.leftChild = TreeNode(key, val, parent=currentNode)
        # 情况2：新key > 当前节点的key → 往右走
        elif key > currentNode.key:
            if currentNode.hasRightChild():
                self._put(key, val, currentNode.rightChild)
            else:
                currentNode.rightChild = TreeNode(key, val, parent=currentNode)
        # 情况3: 新key == 当前节点的key → 更新当前节点的payload
        else:
            currentNode.payload = val

    def __setitem__(self, k, v):
        self.put(k, v)
        # 语法糖，字典原生支持 [] 赋值就是靠__setitem__()
        # 可以写mytree[3] = "red"
        # 而不是只用写mytree.put(3, "red")
    
    def get(self, key):
        if self.root:
            res = self._get(key, self.root)
            if res:
                return res.payload
            else:   # 根也不是
                return None
        else:       # 没有树
            return None

    def _get(self, key, currentNode):
        if not currentNode:
            return None
        elif currentNode.key == key:
            return currentNode
        # 上面俩是给递归留的结束入口
        elif key < currentNode.key:
            return self._get(key, currentNode.leftChild)
        elif key >= currentNode.key:    # 这个等号应该不用担心
            return self._get(key, currentNode.rightChild)
        
    def __getitem(self, key):
        return self.get(key)
    
    def __contains__(self, key):
        if self._get(key, self.root):
            return True
        else:            return False
    # __contains__() 是语法糖，支持 in 操作符，比如 3 in mytree 就会调用 mytree.__contains__(3)
    # 我们可以写 `if 6 in mytree:`
````

接下来是最难的删除一个节点：
- 空树：不用删
- 只有一个根：返回空树
- 删的是叶子节点：`parent.child`直接设None，最多额外加点左右的判断
- 删的节点有一个孩子：子承父业顶上来，`parent.child = parent.child.child`跳过他
- 删的节点有两个孩子——世子之争（因为你不知道到底谁该上位）
  - 我们所需要的是 **后继节点（successor）**，也就是“比当前节点大的最小那个节点”
  - **右子树里最小的节点** 比左边所有节点大，比右边其他节点小或合适
  - 如果节点没有右子节点，且是其父节点的左子节点，那么父节点就是后继节点。
  - 如果一个节点是其父节点的右子节点，且自身没有右子节点，那么该节点的后继节点是其父节点的后继节点（不包括该节点本身）。

![删除节点5，一个有两个子节点的节点](/img/python/bst-del.png)

比如说删50：
```
        50
       /  \
     30    70
          /  \
         60   80
```
- 右子树是以 70 为根
- 右子树最小的是 60
- 所以 60 是 50 的后继
```
        60
       /  \
     30    70
             \
              80
```
删除两个孩子节点时，真正步骤只有 4 步，以删除节点 X 为例：
1. 找到 X 的后继 succ
2. 把 succ 从原位置剪掉
3. 用 succ.key 覆盖 X.key
4. 用 succ.payload 覆盖 X.payload

````python
class TreeNode:
    def findSuccessor(self):
        succ = None
        # 1. 有右子树 → 后继 = 右子树的最小节点
        if self.hasRightChild():
            succ = self.rightChild.findMin()
        # 2. 无右子树：向上追溯
        else:
            current = self
            # 如果是父节点的右子节点，就继续往上找
            while current.parent and current.isRightChild():
                current = current.parent

            # 最终父节点就是后继（根节点无父节点，返回None）
            succ = current.parent
            # else就是根节点，那直接回一个succ = None就行了
        return succ
    
    def findMin(self):
        # 找以当前节点为根的子树的最小节点（一直往左走）
        current = self
        while current.hasLeftChild():
            current = current.leftChild
        return current
    """
        50
       /  \
     30    70
       \
        40
    40的后继：
1. 40没有右子树，向上追溯
2. 40是30的右子节点，继续往上追溯
3. 30是50的左子节点，返回50

也就是说，后继是直到你第一次从左边进入到父节点中
"""
    def spliceOut(self):
        # 后继节点不可能有两个孩子，所以这里只处理 0 个或 1 个孩子
        if self.isLeaf():
            if self.isLeftChild():
                self.parent.leftChild = None
            elif self.isRightChild():
                self.parent.rightChild = None

        else:           # 不是叶子，只有一个孩子
            if self.hasLeftChild():
                if self.isLeftChild():
                    self.parent.leftChild = self.leftChild
                else:
                    self.parent.rightChild = self.leftChild
                self.leftChild.parent = self.parent
            else:
                if self.isLeftChild():
                    self.parent.leftChild = self.rightChild
                else:
                    self.parent.rightChild = self.rightChild
                self.rightChild.parent = self.parent
        
````

最后是遍历，中序遍历。
Python提供了一个在创建迭代器时非常强大的函数yield。yield与return类似，都能向调用者返回一个值。不过，yield还有一个额外的步骤，就是冻结函数的状态，这样下次调用该函数时，它会从之前暂停的精确位置继续执行。创建可迭代对象的函数被称为生成器函数。
- return 一次性结束
- yield 是“吐一个值，暂停住，下次继续”

| 对比维度 | 传统 `inorder` 遍历 | `yield` + 魔法方法 `__iter__` |
| :--- | :--- | :--- |
| **数据存储** | 先把**所有节点**装进列表，再返回 | 一次只生成**一个值**，不建列表 |
| **调用方式** | 必须手动写函数、传列表、调用 | 直接写 `for val in root:`，全自动 |
| **执行逻辑** | 遍历完整棵树，才给你结果 | 要一个值，生成一个值（暂停/继续） |
| **内存占用** | 树越大，内存越占 | 永远只占一个值的内存 |
| **Python风格** | 手动、繁琐 | 优雅、原生语法 |

> 魔法方法 Python 会在你使用对应原生语法时，自动偷偷执行这个方法

````python
class TreeNode:
    def __iter__(self):
        if self:            # 节点非空才执行（避免空节点报错）
            if self.hasLeftChild():
                for element in self.leftChild:
                    yield element
            yield self.key
            if self.hasRightChild():
                for element in self.rightChild:
                    yield element
    TreeNode.__iter__ = __iter__

# 直接for循环遍历根节点，自动中序输出！
# for key in root_node:
#     print(key)
````

````python
# ==============================================
# 【传统中序遍历】→ 收集到列表，一次性return
# ==============================================
def inorder_traversal(node, result):
    if node:
        inorder_traversal(node.leftChild, result)
        result.append(node.key)  # 把值塞进列表
        inorder_traversal(node.rightChild, result)


# 构建树
root = TreeNode(3)
root.leftChild = TreeNode(1)
root.leftChild.rightChild = TreeNode(2)
root.rightChild = TreeNode(4)

print("===== 版本1：传统inorder遍历 =====")
result = []
inorder_traversal(root, result)  # 手动调用函数
for val in result:               # 遍历列表
    print(val)

print("\n===== 版本2：yield+魔法方法 =====")
for val in root:                 # 直接遍历根节点！Python自动调用__iter__
    print(val)
````

## 平衡二叉搜索树 AVL
> 这个名字来源于发明者G.M. Adelson-Velskii and E.M. Landis
它和普通 BST 的区别只有一个 —— AVL 要额外保证“树始终比较平衡”。
`balanceFactor = height(leftSubTree) - height(rightSubTree)`
- balanceFactor > 0：左边更高，左重
- balanceFactor < 0：右边更高，右重
- balanceFactor = 0：两边一样高
AVL 允许的范围是 [-1, 0, 1]，倘若某个节点的平衡因子变成2或-2，说明这个局部已经失衡必须调整。

![](/img/python/avl-performance.png)
AVL 树的高度始终是 O(log n)。
所以：
- 查找 get：O(log n)
- 插入 put：O(log n)
- 删除（如果实现了平衡维护）：O(log n)

![](/img/python/left-rotation.png)
![](/img/python/right-rotation.png)

只看 _put 会发现 AVL 和普通 BST 几乎一样。真正的不同只有一个：插入成功后立刻调用 updateBalance。

````python
def _put(self, key, val, currentNode):
    if key < currentNode.key:
        if currentNode.hasLeftChild():                  # 如果这个节点有左子树，就继续往下走
            self._put(key, val, currentNode.leftChild)  # 递归
        else:
            currentNode.leftChild = TreeNode(key, val, parent=currentNode)  # 没有的话就这新造一个节点
            self.updateBalance(currentNode.leftChild)   # AVL平衡因子
    else:
        if currentNode.hasRightChild():
            self._put(key, val, currentNode.rightChild)
        else:
            currentNode.rightChild = TreeNode(key, val, parent=currentNode)
            self.updateBalance(currentNode.rightChild)

def updateBalance(self, node):
    if node.balanceFactor > 1 or node.balanceFactor < -1:
        # 如果绝对值大于1那么就失衡了，需要旋转
        self.rebalance(node)
        return
    
    if node.parent != None:
        if node.isLeftChild():
            node.parent.balanceFactor = node.parent.balanceFactor + 1
        elif node.isRightChild():
            node.parent.balanceFactor = node.parent.balanceFactor - 1

        # 比较有意思
        # 父节点如果变成 0，说明它这个子树整体高度没有继续增长，所以祖先不会再被影响，递归可以终止。
        if node.parent.balanceFactor != 0: 
            self.updateBalance(node.parent)
````

### 旋转
平衡因子失控后，AVL 不是像链表那样大搬家，也不是重新构建整棵树，而是只做一种很局部的结构调整：**旋转（rotation）**。

也就是说，在不破坏 BST 中序有序性的前提下，重新分配局部父子关系，让高度恢复平衡。
#### 左旋
```
A
 \
  B
   \
    C
```
- 某个节点 **右边太高**；
- 它的平衡因子变成 -2；
这时候以 A 为轴做一次左旋，会变成：
```
    B
   / \
  A   C
```
此时原来的右孩子 B 被提上来，变成新的子树根，原来的根 A 被压到左边去，但这还是太抽象了！它牵扯的子树也应该考虑进来：
```
        A
       / \
      T1  B
         / \
        T2  T3

        B
       / \
      A   T3
     / \
    T1  T2
```
这个 T2 的挪动非常关键，因为它必须保持 BST 的大小关系：
- T2 里的值都大于 A
- 同时又小于 B
- 所以它正好应该成为 A 的右子树


````python
def rotateLeft(self, rotRoot):
    newRoot = rotRoot.rightChild    # rotRoot是A，newRoot是B。动作：newRoot 变量 → 指向 B
    rotRoot.rightChild = newRoot.leftChild  # T2挂到A的右子树上。动作：A.rightChild → 指向 T2
    if newRoot.leftChild != None:
        newRoot.leftChild.parent = rotRoot  # 树是双向指针。动作：如果 T2 存在，那么 T2.parent → 指向 A

    # newRoot 既然要顶替 rotRoot，那它的父亲就该变成 rotRoot 原来的父亲
    newRoot.parent = rotRoot.parent         # A 原来的爸爸是谁，B 现在直接继承

    # 接下来分情况讨论
    # 1. rotRoot 原来就是整颗树根节点，那么 newRoot 就直接变成根了
    if rotRoot.isRoot():
        self.root = newRoot
    # 2. rotRoot 原来是个左子节点，那么就把 rotRoot 的父节点的左指针指向 newRoot
    elif rotRoot.isLeftChild():
        rotRoot.parent.leftChild = newRoot
    elif rotRoot.isRightChild():
        rotRoot.parent.rightChild = newRoot

    # 最后把 rotRoot 降为新根的左孩子
    newRoot.leftChild = rotRoot
    rotRoot.parent = newRoot

    # 旋转时除了重连指针，还要同步修正参与旋转的几个节点的 balanceFactor。
    # 书里给出了常数时间更新公式，重点是理解它的目标：避免每次旋转后重新递归计算整棵子树高度。
    rotRoot.balanceFactor = rotRoot.balanceFactor + 1 - min(newRoot.balanceFactor, 0)
    newRoot.balanceFactor = newRoot.balanceFactor + 1 + max(rotRoot.balanceFactor, 0)
````

#### 右旋
右旋和左旋完全对称：
- 某个节点 **左边太高**
- 平衡因子变成 +2

```
        E (rotRoot 右旋锚点)
       / \
      C   T4
     / \
    T1  D

        E
       / \
  newRoot=C  T4
     / \
    T1  D

        E
       / \
      D   T4   E左边现在挂D
    (C 暂时还飘在外面，只改了E的指针)

        E
       / \
      D   T4
      ↑
      D.parent = E （双向绑定完成）

    然后爹给爹

    最后才完成收官挂靠权力交接
      C (全局新根)
     / \
    T1  E
       / \
      D   T4
```

````python
def rotateRight(self, rotRoot):
    newRoot = rotRoot.leftChild
    rotRoot.leftChild = newRoot.rightChild
    if newRoot.rightChild != None:
        newRoot.rightChild.parent = rotRoot

    newRoot.parent = rotRoot.parent
    if rotRoot.isRoot():
        self.root = newRoot
    else:
        if rotRoot.isLeftChild():
            rotRoot.parent.leftChild = newRoot
        else:
            rotRoot.parent.rightChild = newRoot

    newRoot.rightChild = rotRoot
    rotRoot.parent = newRoot
````

#### 双旋转
左旋对应右右型失衡（RR），右旋对应左左型失衡（LL）。然而有些失衡不是“一条直线偏斜”，而是“拐弯偏斜”：
- 右左型（RL）
    - 某节点整体右重，但它的右孩子却左重
- 左右型（LR）
    - 某节点整体左重，但它的左孩子却右重

这两种情况如果硬做一次单旋，通常会越转越别扭，甚至转完还是不平衡。所以需要 **双旋转**。
```
右左型（RL）：先右旋，再左旋
A
 \
  C
 /
B
```
正确做法是先对 C 右旋、再对 A 左旋。
```
# 右旋C之后，中间过渡形态：
    A
     \
      B
       \
        C

# 最终平衡结果
   B
  / \
 A   C
```

```
左右型（LR）：先左旋，再右旋
    C (祖父)
   /
  A (父) ← 我们要对 **A 执行左旋**
   \
    B (子，A的右孩子)
```
对应地先对 A 左旋、再对 C 右旋：
```
    C
   /
  B   ← B上位了！
 /
A     ← A变成了B的左孩子

   B
  / \
 A   C
```


````python
def rebalnce(self, node):
    if node.balanceFactor < 0:  # 右子树高
        if node.rightChild.balanceFactor > 0:    # RL'
            self.rotateRight(node.rightChild)
            self.rotateLeft(node)
        else:                                   # RR
            self.rotateLeft(node)

    elif node.balanceFactor > 0:    # 左子树高
        if node.leftChild.balanceFactor < 0:     # LR'
            self.rotateLeft(node.leftChild)
            self.rotateRight(node)
        else:                                   # LL
            self.rotateRight(node)
````

| 节点状态 | 子节点状态 | 旋转类型 | 执行操作 |
| :------- | :--------- | :------- | :------- |
| **当前节点左重**<br>`node.balanceFactor > 0` | 左孩子左重<br>`node.leftChild.balanceFactor >= 0` | **LL 型** | 直接**右旋**当前节点 |
| | 左孩子右重<br>`node.leftChild.balanceFactor < 0` | **LR 型** | 先**左旋左孩子**<br>再**右旋当前节点** |
| **当前节点右重**<br>`node.balanceFactor < 0` | 右孩子右重<br>`node.rightChild.balanceFactor <= 0` | **RR 型** | 直接**左旋**当前节点 |
| | 右孩子左重<br>`node.rightChild.balanceFactor > 0` | **RL 型** | 先**右旋右孩子**<br>再**左旋当前节点** |

- 左重左倾 → 右旋（LL）
- 左重右倾 → 先左后右（LR）
- 右重右倾 → 左旋（RR）
- 右重左倾 → 先右后左（RL）

> We leave the deletion of the node and subsequent updating and rebalancing as an exercise for you.


时间复杂度方面：
- 插入路径最长也就是树高
- AVL 的树高始终是 O(log n)
- 沿祖先向上更新最多走 O(log n) 步
- 旋转最多只需要常数次（一次或两次）
- 每次旋转只是改几个指针，属于 O(1)

所以最终：
- get：O(log n)
- put：O(log n)
这就是 AVL 的价值：
**它用更复杂的插入维护，换来了稳定的对数级查询性能**。


| operation | Sorted List | Hash Table | Binary Search Tree | AVL Tree |
|:----------|:-----------:|:----------:|:------------------:|:--------:|
| put       | $O(n)$      | $O(1)$     | $O(n)$             | $O(\log_2 n)$ |
| get       | $O(\log_2 n)$| $O(1)$     | $O(n)$             | $O(\log_2 n)$ |
| in        | $O(\log_2 n)$| $O(1)$     | $O(n)$             | $O(\log_2 n)$ |
| del       | $O(n)$      | $O(1)$     | $O(n)$             | $O(\log_2 n)$ |


## 红黑树
红黑树是带红黑染色规则、靠变色 + 旋转自动维持弱平衡的二叉搜索树，能彻底避免普通 BST 退化成链表，增删查稳定保持 $O(\log_2 n)$，旋转比 AVL 更少，是工程里最常用的平衡树（HashMap、TreeMap 底层核心）
