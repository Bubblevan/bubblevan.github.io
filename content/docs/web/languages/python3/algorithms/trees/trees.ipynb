## 基础特性
1. **层次性**，可沿着“根→中间节点→叶节点”的路径逐层访问；
2. **节点子节点独立性**，不同父节点下的子节点彼此独立，即使名称相同，也属于不同节点，修改其一不会影响另一；
3. **叶节点唯一性**，从树根到任意叶节点的**完整路径**是唯一的，可通过该路径精准标识叶节点；
4. **子树可迁移性**，树的任意子树可整体移动到树的其他位置，仅改变子树的上层路径，不会影响子树内部的层级结构和内容；

### 树的定义

**定义一**	树由一组节点和一组连接节点对的边组成。树具有以下性质：

### 节点与引用
使用一个类去包含根值与左右子树属性的方法更贴近OOP范式。

````python
class BinaryTree:
    def __init__(self, rootObj):
        self.key = rootObj
        self.leftChild = None
        self.rightChild = None

    def insertLeft(self, newNode):
        if self.leftChild == None:
            self.leftChild = BinaryTree(newNode)
        else:   # 左子树存在的话
            t = BinaryTree(newNode)
            t.leftChild = self.leftChild
            self.leftChild = t
            # 不是左子树的左子树（递归）的意思！
            # 而是左子树往下挪，腾出位置插新节点

    def insertRight(self, newNode):
        if self.rightChild == None:
            self.rightChild = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.rightChild = self.rightChild
            self.rightChild = t

    def getRightChild(self):
        return self.rightChild
    def getLeftChild(self):
        return self.leftChild
    def setRootVal(self, obj):
        self.key = obj
    def getRootVal(self):
        return self.key

r = BinaryTree('a')
print(r.getRootVal())
print(r.getLeftChild())
r.insertLeft('b')
print(r.getLeftChild())
print(r.getLeftChild().getRootVal())
r.insertRight('c')
print(r.getRightChild())
print(r.getRightChild().getRootVal())
r.getRightChild().setRootVal('hello')
print(r.getRightChild().getRootVal())
````

#### 从算式到解析树
`( 3 + ( 4 * 5 ) )`拆成列表就是
```
['(', '3', '+', '(', '4', '*', '5', ')', ')']
```
分为4种规则：
- 左括号`(`：给当前节点加一个左孩子，然后往左走到这个左孩子
- 运算符`+-*/`：把当前节点的值设为这个运算符，然后加一个右孩子并往右走
- 数字：当前节点值设为该数字，回到父节点（准备读运算符了嘛）
- 右括号`)`：回到父节点（子表达式结束）


````python


def buildParseTree(fpexp):
    fplist = fpexp.split()  # 把表达式拆成token
    pStack = Stack()        # 用栈记父节点
    eTree = BinaryTree('')

    pStack.push(eTree)
    currentTree = eTree

    for i in fplist:
        if i == '(':
            currentTree.insertLeft('')
            pStack.push(currentTree)
            currentTree = currentTree.getLeftChild()

        elif i in ['+', '-', '*', '/']:
            currentTree.setRootVal(i)
            currentTree.insertRight('')
            pStack.push(currentTree)
            currentTree = currentTree.getRightChild()

        elif i == ')':
            currentTree = pStack.pop()

        elif i not in ['+', '-', '*', '/', ')']:
            try:
                currentTree.setRootVal(int(i))
                parent = pStack.pop()
                currentTree = parent

            except ValueError:
                raise ValueError("token '{}' is not a valid integer".format(i))

    return eTree

pt = buildParseTree("( ( 10 + 5 ) * 3 )")
pt.postorder()  #defined and explained in the next section
````

#### 从解析树到结果
递归思想

````python
import operator

def evaluate(parseTree):
    opers = {
        '+': operator.add,
        '-': operator.sub,
        '*': operator.mul,
        '/': operator.truediv
    }

    leftC = parseTree.getLeftChild()
    rightC = parseTree.getRightChild()

    if leftC and rightC:
        fn = opers[parseTree.getRootVal()]
        return fn(evaluate(leftC), evaluate(rightC))
    else:
        return parseTree.getRootVal()
````

### 优先队列
普通队列是FIFO，但是优先队列则是按**优先级**决定**出队顺序**。
比如说用列表去实现这样的优先队列，插入元素的时间复杂度是O(n)，再进行排序的时间复杂度是O(nlogn)。
有一说一，超级麻烦：
- 插入后还要找位置
- 或者每次取最小值都要重新排序
但实际上可以用一种称为**二叉堆**的数据结构做得更好——其能够实现出队和入队操作的时间复杂度都达到O(logn)！！
## 二叉堆
### 分类
- 小顶堆：每个节点的值都小于等于其左右子节点的值
- 大顶堆：每个节点的值都大于等于其左右子节点的值
### 实现小顶堆
#### 结构特性
完全二叉树：除了最后一层，前面每层都满。最后一层从左到右依次填。
![](/img/python/complete-binary-tree.png)
用完全二叉树可以保证节点位置的规整——能直接用列表表示。
加入一个节点在下标i：
- 左孩子在2i
- 右孩子在2i+1
- 爸爸在i//2
#### 堆序特性
除了“形状上是完全二叉树”，堆序特性要求（最小堆）满足**每个父节点都 小于等于 它的孩子**：
- 根节点一定最小
- 但整棵树不要求**完全排序**，他只要求爸爸<=儿子，对世子之争没有要求
```
        2
      /   \
     3     5
    / \
   9   6
```


````python
class BinHeap:
    def __init__(self):
        self.heapList = [0]   # heapList 用来存堆，第 0 位放个没用的占位符
        self.currentSize = 0  # currentSize 记录现在有几个元素
````

![将新元素在树中向上渗透到维持堆特性所需的位置](/img/python/percolate-up.png)
一旦新元素被添加到树中，percUp就会接管并将新元素放置在合适的位置。

````python
class BinHeap:
    def percUp(self, i):
        while i // 2 > 0:
            if self.heapList[i] < self.heapList[i // 2]:
                tmp = self.heapList[i//2]               # 获取父节点，值是tmp
                self.heapList[i//2] = self.heapList[i]  # 被插入值移到上面去
                self.heapList[i] = tmp                  # 爸爸下来
            i = i // 2

    def insert(self, k):
        self.heapList.append(k)
        self.currentSize = self.currentSize + 1
        self.percUp(self.currentSize)
````

![](/img/python/delmin.png)
因为最小堆的根节点最小，所以删最小值就是删根。但删根之后不能空着，需要做两步来**恢复堆结构**：
- 把最后一个元素搬到根
- 让它往**下沉**，恢复堆序性质

> 要理解下面的下沉方法的话，需要从下往上看这3个方法

````python
class BinHeap:
    def minChild(self, i):
        if i*2 + 1 > self.currentSize:
            return i*2        # 没有右孩子那就是左孩子
        else:
            if self.heapList[i*2] < self.heapList[i*2+1]:
                return i*2
            else:
                return i*2+1

    def percDown(self, i):
        while(i*2) <= self.currentSize:
            mc = self.minChild(i)             # mc应该是minimal character的意思？总而言之需要左右世子中更小的那个，因此自己实现
            if self.heapList[i] > self.heapList[mc]:
                self.heapList[i], self.heapList[mc] = self.heapList[mc], self.heapList[i]
            i = mc

    def delMin(self):
        retval = self.heapList[1]
        self.heapList[1] = self.heapList[self.currentSize]  # 把最后一个元素搬到根
        self.currentSize = self.currentSize - 1
        self.heapList.pop()  # 删除最后一个元素
        self.percDown(1)    # 让它往下沉，恢复堆序性质
        return retval
````

给你一个普通列表，如何直接原地整理成一个堆，而不是一个个调用insert？
![](/img/python/buildheap.png)

````python
class BinHeap:
    def buildHeap(self, alist):
        i = len(alist) // 2       # 因为最小的左叶子是2i嘛，后面的都是叶子，所以根节点只能在1~2i之间
        self.currentSize = len(alist)
        self.heapList = [0] + alist[:]
        while(i > 0):             # 从“最后一个非叶子节点”开始，一路往前下沉到叶子，直到根
            self.percDown(i)
            i = i - 1
````

## 二叉搜索树
前面用完全二叉树->最小堆/最大堆，是为了实现优先队列；
而这里是用二叉搜索树来实现字典这种**键值对结构**。
map 本质是**用 key 找 value**。给定`{6, "yellow"}`的6和一个字典，我如何最少的查询次数来得到"yellow"？
查找时不是遍历全部，而是根据大小关系缩小范围：
- 如果目标 key 比当前小 → 去左边
- 比当前大 → 去右边
- 相等 → 找到了
这就像**“树形版二分查找”**。根据 key 的大小关系，快速找到对应的 value。
BST 的核心性质就一句：**左边比我小，右边比我大**。
这个规则不只是根节点成立，是**每个节点**都成立。
